```python
def get_Quater(month):
    if month==1 or month==2 or month==3:
       return 'Q1'
    elif month==4 or month==5 or month==6:
       return 'Q2'
    elif month==7 or month==8 or month==9:
       return 'Q3'
    elif month==10 or month==11 or month==12:
       return 'Q4'
    else:
      return ''
```
